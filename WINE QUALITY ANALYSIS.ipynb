```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
import numpy as np

X = data[:, :11]
Y = data[:, -1]

# Check unique values in the target variable
print("Unique values in Y:", np.unique(Y))

# Convert Y to categorical (if needed)
# Example: Binarize assuming a threshold of 0.5
Y_binary = (Y > 0.5).astype(int)

# Splitting the dataset into training and test set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y_binary, test_size=0.25, random_state=0)

# Feature scaling
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# Training using Logistic Regression
cl = LogisticRegression()
cl.fit(X_train, Y_train)

# Making predictions
Y_pred = cl.predict(X_test)

# Making confusion matrix
cm = confusion_matrix(Y_test, Y_pred)
print(cm)
```

```python
from sklearn.svm import SVC
cl = SVC(kernel="rbf")
cl.fit(X_train,Y_train)

cm = confusion_matrix(Y_test,cl.predict(X_test))
print(cm)
```
